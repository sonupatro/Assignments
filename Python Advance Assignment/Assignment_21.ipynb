Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you
predict them at all?

Ans :

A probability distribution is a function that describes the likelihood of obtaining the possible values that a random variable can take. It assigns probabilities to each of these values, indicating how often we can expect to see them in repeated trials of an experiment. Although the values are meant to be random, a probability distribution can still provide information about their expected behavior, such as the average or most likely outcome. It is important to note that the actual values observed will vary each time the experiment is conducted, but the probability distribution provides a basis for predicting the overall pattern of results.

Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is
one? Why are the latter considered “good enough”?

Ans :
    
Yes, there is a distinction between true random numbers and pseudo-random numbers. True random numbers are generated based on unpredictable physical phenomena, such as atmospheric noise or radioactive decay. Pseudo-random numbers, on the other hand, are generated by algorithms that use mathematical formulas to produce sequences of numbers that appear random, but are actually deterministic.

Pseudo-random numbers are considered "good enough" because they are computationally efficient and can produce sequences of numbers that are statistically random. In many applications, such as simulations or cryptography, the randomness of the numbers generated is not as important as their unpredictability, which pseudo-random numbers can provide. Additionally, generating truly random numbers can be expensive and time-consuming, whereas pseudo-random numbers can be generated quickly and efficiently.

Q3. What are the two main factors that influence the behaviour of a &quot;normal&quot; probability distribution?

Ans :
    
The two main factors that influence the behavior of a normal (Gaussian) probability distribution are the mean (μ) and the standard deviation (σ).

The mean represents the center or the average of the distribution, and determines the position of the bell-shaped curve. If the mean shifts to the left or right, the entire distribution shifts along with it.

The standard deviation represents the spread or the variability of the distribution, and determines the shape of the bell-shaped curve. A smaller standard deviation means that the data is more tightly clustered around the mean, while a larger standard deviation means that the data is more spread out. If the standard deviation increases, the bell-shaped curve becomes wider, and if it decreases, the curve becomes narrower.

Q5. In the short term, how can you expect a probability distribution to behave? What do you think will
happen as the number of trials grows?

Ans :

In the short term, the behavior of a probability distribution can be highly variable and may not reflect the long-term behavior of the distribution. For example, in a game of chance with a fair coin, you could expect to get heads 50% of the time in the long term, but in a small number of flips, you could get heads or tails much more or much less frequently.

However, as the number of trials grows, the distribution will converge to its expected behavior. For example, in the coin flipping example, if you were to flip the coin a large number of times, the frequency of heads would approach 50% and the variability of the results would decrease. This is known as the law of large numbers, which states that the average of a large number of independent trials will approach the expected value.

In general, as the number of trials grows, the distribution will become more stable and predictable, and its behavior will become closer to the expected behavior described by its underlying probability distribution.
